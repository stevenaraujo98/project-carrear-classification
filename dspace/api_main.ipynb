```python
data = {'NOMBREPROYEC': [], 'URL': [], 'CARRERA': [], 'FACULTAD': []}#, 'FECHA': []}
```

```python
for handle_url in list_carreras:
    url = base_url + handle_url + param_url
    response = requests.get(url)
    # len(response)
    
    if response.status_code == 200:
        data_json = response.json()

        # carrera = response["name"]
        carrera = data_json["shortDescription"]

        items = data_json['items']
        count = data_json['numberItems']
        print(carrera, "-", count, len(items) == count)

        counter = 1
        for item in items:
            # print(counter, end=' ')
            title = item['name']
            # print(title)
            url_item = "https://dspace.espol.edu.ec/handle/" + item['handle']
            
            data['NOMBREPROYEC'].append(title)
            data['URL'].append(url_item)
            data['CARRERA'].append(carrera)
            data['FACULTAD'].append(facultad)
            # data['FECHA'].append(get_anio(item['handle']))
            counter += 1
        print(counter, "items completados")
        print("-" * 100)
```

```python
df = pd.DataFrame.from_dict(data)
df.to_csv('results/' + abreviatura + '/dspace_data_' + "_".join(facultad.lower().split()) + '.csv', index=False)
```
